This is a classic dataset that contains four measurements for 150 different [iris flowers](https://en.wikipedia.org/wiki/Iris_(plant)).  Three species are represented, 50 flowers from each.  There are no missing values. The following fields are present:

* **mpg** - The [miles per gallon](https://en.wikipedia.org/wiki/Fuel_economy_in_automobiles) for the car.
* **cylinders** - The number of [cylinders](https://en.wikipedia.org/wiki/Cylinder_(engine)) in the car.
* **displacement** - The engine [displacement](https://en.wikipedia.org/wiki/Engine_displacement).
* **horsepower** - The [horsepower](https://en.wikipedia.org/wiki/Horsepower) produced by the car/engine.	
* **weight** - The weight of the car.
* **acceleration** - The acceleration of the car.
* **year** - The year the car was produced.
* **origin** - Where the car was produced, 1=USA, 2=Europe, 3=Asia.
* **name** - The name of the car.
